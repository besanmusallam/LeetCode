```python
class Solution(object):
    def mergeAlternately(self,word1, word2):
        word=""
        n=min(len(word1),len(word2))
        for i in range(n):
            word+=word1[i]+word2[i]
        if len(word1)>n:
            word+=word1[n:]
        elif len(word2)>n:
            word+=word2[n:]
        return word


word1=input()
word2=input()
ret = Solution().mergeAlternately(word1, word2)
print(ret)
```

```python
class Solution(object):
    def gcdOfStrings(self, str1, str2):
        a=[]
        count=0
        for i in range(len(str2)):
            if str2[i] in str1:
                a.append(str2[i])
            elif str2[i] not in str1:
                count+=1
        if count==0:

            a="".join(a)
            return a
        else:
            return '""'
word1 = input()
word2 = input()
ret = Solution().gcdOfStrings(word1, word2)
print(ret)
```

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        lst=[]
        max_candies = max(candies)
        for i in range(len(candies)):
            if (candies[i]+ extraCandies)>=max_candies:
                lst.append(True)
            else:
                lst.append(False)
        return lst

candies = input()
input_list = list(map(int, candies.split()))
extraCandies = int(input())
ret = Solution().kidsWithCandies(input_list, extraCandies)
print(ret)
```
