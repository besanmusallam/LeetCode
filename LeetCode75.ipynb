```python
class Solution(object):
    def mergeAlternately(self,word1, word2):
        word=""
        n=min(len(word1),len(word2))
        for i in range(n):
            word+=word1[i]+word2[i]
        if len(word1)>n:
            word+=word1[n:]
        elif len(word2)>n:
            word+=word2[n:]
        return word


word1=input()
word2=input()
ret = Solution().mergeAlternately(word1, word2)
print(ret)
```

```python
class Solution(object):
    def gcdOfStrings(self, str1, str2):
        a=[]
        count=0
        for i in range(len(str2)):
            if str2[i] in str1:
                a.append(str2[i])
            elif str2[i] not in str1:
                count+=1
        if count==0:

            a="".join(a)
            return a
        else:
            return '""'
word1 = input()
word2 = input()
ret = Solution().gcdOfStrings(word1, word2)
print(ret)
```

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        lst=[]
        max_candies = max(candies)
        for i in range(len(candies)):
            if (candies[i]+ extraCandies)>=max_candies:
                lst.append(True)
            else:
                lst.append(False)
        return lst

candies = input()
input_list = list(map(int, candies.split()))
extraCandies = int(input())
ret = Solution().kidsWithCandies(input_list, extraCandies)
print(ret)
```

```python
class Solution(object):
    def canPlaceFlowers(self, flowerbed, n):
        count_value = flowerbed.count(0)
        consecutive_zeros = 0
        for num in flowerbed:
            if num == 0:
                 consecutive_zeros  += 1
                 if consecutive_zeros >= 2 * n + 1:
                     return True

            else:
                consecutive_zeros = 0
        if consecutive_zeros >= n:
            return True
        return False



flower= input()
flowerbed= list(map(int, flower.split()))
n=int(input())
ret = Solution().canPlaceFlowers(flowerbed, n)
print(ret)
```

```python
class Solution(object):
    def reverseVowels(self, s):
        vowels=['a','e','i','o','u','A','E','I','O','U']
        s = list(s)
        lst=[]
        num=[]
        for i in range(len(s)):
            if s[i] in vowels:
                lst.append(s[i])
                num.append(i)
        for i in range(len(num)):
            for j in range(len(lst) - 1, -1, -1):
                s[num[i]]= lst[j]
                lst.pop()
                break
        return ''.join(s)

s= input()
ret = Solution().reverseVowels(s)
print(ret)
```
